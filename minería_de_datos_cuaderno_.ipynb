```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/doc.csv')
df.columns = [c.strip() for c in df.columns]

# --- Variables derivadas ---
df['edad'] = pd.to_numeric(df['쮺u치ntos a침os cumplidos tienes?'], errors='coerce')
map_sn = {'Si': True, 'No': False}
df['prog_python'] = df['쮿as realizado programas utilizando Python?'].map(map_sn)
df['usa_colab'] = df['쮿as utilizado Google Colab?'].map(map_sn)
df['cuenta_github'] = df['쯊ienes cuenta de GitHub?'].map(map_sn)
```

```python
# --- Paleta de colores ---
accent  = '#2563eb'  # azul
accent2 = '#16a34a'  # verde
accent3 = '#f59e0b'  # 치mbar
neutral = '#6b7280'  # gris
```

```python
import matplotlib.pyplot as plt

# 1) Histograma de edades
# =========================
fig, ax = plt.subplots(figsize=(8,5))
bins = range(int(df['edad'].min())-1, int(df['edad'].max())+2)
ax.hist(df['edad'], bins=bins, color=accent, edgecolor='white', alpha=0.85)
mean_age = df['edad'].mean()
med_age = df['edad'].median()
ax.axvline(mean_age, color=accent2, linestyle='--', linewidth=2, label=f"Media: {mean_age:.2f}")
ax.axvline(med_age, color=accent3, linestyle=':', linewidth=2.5, label=f"Mediana: {med_age:.2f}")
ax.set_title('Distribuci칩n de edades')
ax.set_xlabel('Edad (a침os)')
ax.set_ylabel('Frecuencia')
ax.legend()
plt.tight_layout()
plt.show()
fig.savefig('01_hist_edades.png', dpi=180)
```

```python
# 2) Boxplot de edades
# =======================
fig, ax = plt.subplots(figsize=(6,4))
ax.boxplot(df['edad'], vert=False, patch_artist=True,
           boxprops=dict(facecolor=accent, color=accent),
           medianprops=dict(color='white', linewidth=2),
           whiskerprops=dict(color=accent), capprops=dict(color=accent))
ax.set_title('Caja y bigotes: edad')
ax.set_xlabel('Edad (a침os)')
plt.tight_layout()
plt.show()
fig.savefig('02_box_edades.png', dpi=180)
```

```python
# ====================================================
# 3) Barras: frecuencias de S칤/No para Python/Colab/GitHub
# ====================================================
fig, axes = plt.subplots(1,3, figsize=(12,4), sharey=True)
for ax, col, title, color in zip(
    axes,
    ['prog_python','usa_colab','cuenta_github'],
    ['Program칩 en Python','Us칩 Google Colab','Cuenta de GitHub'],
    [accent, accent2, accent3]
):
    counts = df[col].value_counts().reindex([True, False]).fillna(0).astype(int)
    labels = ['S칤','No']
    ax.bar(labels, counts, color=[color, neutral], edgecolor='white')
    ax.set_title(title)
    ax.set_ylim(0, max(counts.max(), 1)+2)
    for i, v in enumerate(counts):
        ax.text(i, v+0.1, str(v), ha='center', va='bottom', fontsize=9)
axes[0].set_ylabel('N칰mero de estudiantes')
plt.tight_layout()
plt.show()
fig.savefig('03_barras_frecuencias.png', dpi=180)
```

```python
# 4) Mapas de calor 2x2 para combinaciones
# =======================================
pairs = [
    ('prog_python','usa_colab','Python vs Colab'),
    ('prog_python','cuenta_github','Python vs GitHub'),
    ('usa_colab','cuenta_github','Colab vs GitHub')
]

for idx, (a,b,titulo) in enumerate(pairs, start=1):
    a_bool = df[a].to_numpy(dtype=bool)
    b_bool = df[b].to_numpy(dtype=bool)
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a_bool & b_bool)
    t[1,0] = np.count_nonzero(a_bool & ~b_bool)
    t[0,1] = np.count_nonzero(~a_bool & b_bool)
    t[0,0] = np.count_nonzero(~a_bool & ~b_bool)

    fig, ax = plt.subplots(figsize=(5,4))
    im = ax.imshow(t, cmap='Blues')
    ax.set_title(f'Tabla 2칑2: {titulo}')
    ax.set_xticks([0,1]); ax.set_xticklabels(['No','S칤'])
    ax.set_yticks([0,1]); ax.set_yticklabels(['No','S칤'])
    ax.set_xlabel(b); ax.set_ylabel(a)

    for i in range(2):
        for j in range(2):
            ax.text(j, i, t[i,j], ha='center', va='center',
                    color='black', fontsize=12, fontweight='bold')

    cbar = fig.colorbar(im, ax=ax)
    cbar.ax.set_ylabel('Conteo', rotation=270, labelpad=12)
    plt.tight_layout()
    plt.show()
    fig.savefig(f'04_heatmap_{idx}.png', dpi=180)
```

```python
# 5) Barras: edad promedio por grupo (S칤/No) por variable
# ======================================================
fig, axes = plt.subplots(1,3, figsize=(12,4), sharey=True)
for ax, col, title, color in zip(
    axes,
    ['prog_python','usa_colab','cuenta_github'],
    ['Edad promedio por Python','Edad promedio por Colab','Edad promedio por GitHub'],
    [accent, accent2, accent3]
):
    means = df.groupby(col)['edad'].mean().reindex([True, False])
    labels = ['S칤','No']
    bars = ax.bar(labels, means, color=[color, neutral], edgecolor='white')
    for i, v in enumerate(means):
        ax.text(i, v+0.05, f"{v:.2f}", ha='center', va='bottom', fontsize=9)
    ax.set_title(title)
    ax.set_xlabel('Grupo')
axes[0].set_ylabel('Edad promedio (a침os)')
plt.tight_layout()
plt.show()
fig.savefig('05_barras_edad_promedio.png', dpi=180)
```

####MSE

```python
import numpy as np
import pandas as pd
```

```python
anios_escolaridad = np.array([0, 3, 6, 9, 12, 16, 18])
ingreso_mensual_estimado_mx = np.array([12621, 15723, 19587, 24401, 30398, 40747, 47176])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx)
plt.title('A침os de Escolaridad vs. Ingreso Mensual Estimado')
plt.xlabel('A침os de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.show()
```

Primer intento de modelar los a침os de escolaridad y el ingreso estimado.   
Utilizar el primer punto y el 칰tlimo para trazar una recta:   
$$ m = \frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

$$ b = y_{1} - m* x_{1}$$


```python
x1 = anios_escolaridad[0]
y1 = ingreso_mensual_estimado_mx[0]
x2 = anios_escolaridad[-1]
y2 = ingreso_mensual_estimado_mx[-1]

m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

print(f"La ecuaci칩n de la recta es: y = {m:.2f}x + {b:.2f}")
```

```python

import matplotlib.pyplot as plt

# Calcular los valores y de la recta utilizando m y b
linea_ingreso_estimado_new_plot = m * anios_escolaridad + b

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx, label='Datos Originales')
plt.plot(anios_escolaridad, linea_ingreso_estimado_new_plot, color='red',
         label=f'Recta de Regresi칩n (y = {m:.2f}x + {b:.2f})')
plt.title('A침os de Escolaridad vs. Ingreso Mensual Estimado con L칤nea de Regresi칩n')
plt.xlabel('A침os de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.legend()
plt.show()
```

Valores predichos $$ y(_{pred})$$ : Son los ingresos mensuales que nuestra recta de regresi칩n predice para cada a침o de escolaridad.

Errores (residuales): Es la diferencia entre el ingreso mensual real (ingreso_mensual_estimado_mx) y el ingreso mensual predicho (y_pred). Un error positivo significa que la predicci칩n fue menor que el valor real, y un error negativo indica que la predicci칩n fue mayor.

Errores al cuadrado: Hemos elevado al cuadrado cada uno de los errores. Esto se hace para eliminar los signos negativos (de modo que los errores positivos y negativos no se cancelen entre s칤) y para penalizar m치s los errores grandes.

Suma de errores cuadrados: Es la suma de todos los errores al cuadrado. Este valor nos da una medida total de la 'distancia' de nuestros puntos a la l칤nea.

Error Cuadr치tico Medio (MSE): Es la suma de los errores cuadrados dividida por el n칰mero total de puntos de datos. El MSE es una medida promedio de la magnitud de los errores. Un MSE m치s bajo indica que el modelo se ajusta mejor a los datos.

En este caso, el Error Cuadr치tico Medio (MSE) es de aproximadamente 13,196,667.31. Este n칰mero nos da una idea del rendimiento general de la recta que trazamos, mostrando la magnitud promedio de los errores al cuadrado entre los valores observados y los valores predichos por la l칤nea.

F칩rmula del MSE (Error Cuadr치tico Medio) del Ingl칠s Mean Squared Error   

$$ MSE = \frac{\Sigma_{i=1}^{n}(Y_{i}-\hat{Y_{i}})^{2}}{n}$$

Donde:   
$MSE =$ error cuadr치tico medio.  
$n= $n칰mero de datos.   
$Y_{i}=$ valores observados.   
$ \hat{Y_{i}}$valores predichos.



```python
# 1. Calcular los valores predichos (y_pred) usando la ecuaci칩n de la recta
y_pred = m * anios_escolaridad + b
y_pred
```

```python
# 2. Calcular los errores (residuales): la diferencia entre los valores reales y los predichos
errores = ingreso_mensual_estimado_mx - y_pred
errores
```

```python
# 3. Elevar al cuadrado cada error
errores_cuadrados = errores**2
errores_cuadrados
```

```python
# 4. Sumar los errores cuadrados
suma_errores_cuadrados = np.sum(errores_cuadrados)
suma_errores_cuadrados
```

```python
# 5. Dividir la suma de los errores cuadrados por el n칰mero de puntos para obtener el MSE
mse = suma_errores_cuadrados / len(anios_escolaridad)
mse
```
