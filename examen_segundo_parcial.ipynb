```python
import pandas as pd

dfc = pd.read_csv("clientes.csv")   # clientes
dfo = pd.read_csv("ordenes.csv")    # órdenes

df_merged = dfc.merge(dfo, on="cliente_id", how="left")
num_filas = len(df_merged)
print("Número de filas:", num_filas)
monto_por_cliente = (
    df_merged
    .groupby("cliente_id")["monto"]
    .sum()
    .fillna(0)
)

print("Montos por cliente:")
print(monto_por_cliente)

monto_cliente_1 = monto_por_cliente.loc[1]
print("Monto total del cliente 1:", monto_cliente_1)
```

pregunta 4

```python
import pandas as pd

df = pd.read_csv("sensores.csv")[["ts", "temperatura", "humedad"]]

df["ts"] = pd.to_datetime(df["ts"], format="mixed", errors="coerce")
num_nat = df["ts"].isna().sum()

mediana_temp = df["temperatura"].median()
media_hum = df["humedad"].mean()

df["temperatura"] = df["temperatura"].fillna(mediana_temp)
df["humedad"] = df["humedad"].fillna(media_hum)

mediana_temp_final = round(df["temperatura"].median(), 1)
media_hum_final = round(df["humedad"].mean(), 3)

print(num_nat)
print(mediana_temp_final)
print (media_hum_final)
```
